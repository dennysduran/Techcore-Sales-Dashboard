```python
import pandas as pd
```

```python
df = pd.read_csv("VentasTransformed.csv", encoding="cp1252")
df.head()
```

```python
df.columns
```

```python
df['FechaVenta'] = pd.to_datetime({
    'year': df['Año'],
    'month': df['Mes'],
    'day': df['Día']
})
df = df.rename(columns={'VentaID': 'FacturaID'})
```

## TABLA CIUDADES

```python
ciudades= df['CiudadSucursal'].unique()
ciudades= pd.DataFrame(ciudades, columns=['Ciudad'])
ciudades['CiudadID'] = ciudades.index + 1
ciudades = ciudades[['Ciudad','CiudadID']]
ciudades.head()

```

## TABLA SUCURSALES


```python
sucursales = df[['SucursalNombre', 'CiudadSucursal']].drop_duplicates().reset_index(drop=True)
sucursales['SucursalID'] = sucursales.index + 1
sucursales = sucursales.merge(ciudades, left_on='CiudadSucursal', right_on='Ciudad', how='left')
sucursales = sucursales [['SucursalID','SucursalNombre','CiudadID']]
sucursales
```

## TABLA VENDEDORES


```python
vendedores = df['VendedorNombre'].unique()
vendedores = pd.DataFrame(vendedores, columns=['VendedorNombre'])
vendedores ['VendedorID'] = vendedores.index +1
vendedores = vendedores[['VendedorID','VendedorNombre']]

vendedores
```

## TABLA CLIENTES


```python
Clientes = df[['ClienteNombre','GeneroCliente','EdadCliente','EmailCliente',
               'TelefonoCliente','DireccionCliente']].drop_duplicates().reset_index(drop=True)
Clientes['ClienteID'] = Clientes.index +1
Clientes = Clientes[['ClienteID','ClienteNombre','GeneroCliente','EdadCliente','EmailCliente',
               'TelefonoCliente','DireccionCliente']]


Clientes.head()
```

## TABLA DE MÉTODOS DE PAGO


```python
MetodosPago = df['MetodoPago'].unique()
MetodosPago = pd.DataFrame(MetodosPago, columns=['MetodoPago'])
MetodosPago['MetodoPagoID'] = MetodosPago.index +1
MetodosPago = MetodosPago[['MetodoPago','MetodoPagoID']]



MetodosPago
```

## TABLA FACTURAS


```python
Facturas = df[['FacturaID','FechaVenta','HoraVenta','SucursalNombre','VendedorNombre',
               'ClienteNombre','EmailCliente','MetodoPago','DescuentoVenta','TotalVenta']]
Facturas = Facturas.merge(sucursales, left_on='SucursalNombre' , right_on='SucursalNombre', how= 'left')
Facturas = Facturas.merge(vendedores, left_on='VendedorNombre' , right_on='VendedorNombre', how= 'left')
Facturas = Facturas.merge(Clientes, left_on=['ClienteNombre','EmailCliente'] , right_on=['ClienteNombre','EmailCliente'], how= 'left')
Facturas = Facturas.merge(MetodosPago, left_on='MetodoPago' , right_on='MetodoPago', how= 'left')
Facturas = Facturas[['FacturaID','FechaVenta','HoraVenta','SucursalID','VendedorID',
                     'ClienteID','MetodoPagoID','DescuentoVenta','TotalVenta']]


Facturas.head(20)
```

```python
print(Facturas[Facturas['DescuentoVenta'] > 0]['DescuentoVenta'].unique())
```

## TABLA DE PRODUCTOS

```python
Productos = pd.DataFrame(columns=['NombreProducto','MarcaProducto','Precio'])
for x in range(1,4):
    append_producto = pd.DataFrame(columns=['NombreProducto','MarcaProducto','Precio'])
    append_producto ['NombreProducto'] = df[f'NombreProducto{x}']
    append_producto ['MarcaProducto']  = df[f'MarcaProducto{x}']
    append_producto ['Precio']         = df[f'PrecioUnitarioProducto{x}']
    Productos = pd.concat([Productos,append_producto],ignore_index=True)
    
Productos = Productos.dropna().drop_duplicates().reset_index(drop=True)
Productos  ['ProductoID'] = Productos.index + 1
Productos = Productos[['ProductoID','NombreProducto','MarcaProducto','Precio']]

Productos.head()
```

## TABLA DETALLE FACTURAS

```python
lista_detalles = []

for i in range(1, 4):
    detalle_temp = df[['FacturaID', f'NombreProducto{i}', f'MarcaProducto{i}', f'CantidadProducto{i}']].copy()
    detalle_temp = detalle_temp.merge(
        Productos,
        left_on=[f'NombreProducto{i}', f'MarcaProducto{i}'],
        right_on=['NombreProducto', 'MarcaProducto'],
        how='left'
    )
    detalle_temp = detalle_temp[['FacturaID', 'ProductoID', f'CantidadProducto{i}']].rename(columns={f'CantidadProducto{i}': 'Cantidad'})
    lista_detalles.append(detalle_temp)

DetalleFacturas = pd.concat(lista_detalles, ignore_index=True)
DetalleFacturas = DetalleFacturas.dropna()
DetalleFacturas = DetalleFacturas[DetalleFacturas['Cantidad'] > 0]
DetalleFacturas['FacturaID'] = DetalleFacturas['FacturaID'].astype(int)
DetalleFacturas['ProductoID'] = DetalleFacturas['ProductoID'].astype(int)
DetalleFacturas['Cantidad'] = DetalleFacturas['Cantidad'].astype(int)
DetalleFacturas = DetalleFacturas.sort_values(by=['FacturaID']).reset_index(drop=True)
DetalleFacturas.insert(0, 'DetalleID', DetalleFacturas.index + 1)

print(DetalleFacturas['FacturaID'].unique()[:10])
DetalleFacturas.head(5)
```

## HUERFANOS

```python
h1 = DetalleFacturas[~DetalleFacturas['FacturaID'].isin(Facturas['FacturaID'])]
print(f'FacturaIDs      huérfanos en DetalleFacturas: {len(h1)}')
h2 = DetalleFacturas[~DetalleFacturas['ProductoID'].isin(Productos['ProductoID'])]
print(f'ProductoIDs     huérfanos en DetalleFacturas: {len(h2)}')
h3 = Facturas[~Facturas['SucursalID'].isin(sucursales['SucursalID'])]
print(f'SucursalIDs     huérfanos en DetalleFacturas: {len(h3)}')
h4 = Facturas[~Facturas['VendedorID'].isin(vendedores['VendedorID'])]
print(f'VendedorIDs     huérfanos en DetalleFacturas: {len(h4)}')
h5 = Facturas[~Facturas['ClienteID'].isin(Clientes['ClienteID'])]
print(f'ClienteIDs      huérfanos en DetalleFacturas: {len(h5)}')
h6 = Facturas[~Facturas['MetodoPagoID'].isin(MetodosPago['MetodoPagoID'])]
print(f'MetodoPagoIDs   huérfanos en DetalleFacturas: {len(h6)}')
h7 = sucursales[~sucursales['CiudadID'].isin(sucursales['CiudadID'])]
print(f'CiudadIDs       huérfanos en DetalleFacturas: {len(h7)}')
```

```python
VentasMarca = DetalleFacturas.merge(Productos, on='ProductoID')[['MarcaProducto','Subtotal']].groupby('MarcaProducto').sum().sort_values('Subtotal',ascending=False)
pd.options.display.float_format = '{:,.0f}'.format
print(VentasMarca)
```

```python
TopVentas = DetalleFacturas.merge(Productos, on='ProductoID')[['NombreProducto','Cantidad']].groupby('NombreProducto').sum().sort_values('Cantidad',ascending=False).head(10)
pd.options.display.float_format = '{:,.0f}'.format
TopVentas
```
